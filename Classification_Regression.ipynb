```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import (
    mean_absolute_error, mean_squared_error, r2_score,
    accuracy_score, confusion_matrix, classification_report
)
import joblib

# Verify Versions
print(f"✅ Pandas: {pd.__version__}")
print(f"✅ NumPy: {np.__version__}")
print(f"✅ Scikit-learn: {__import__('sklearn').__version__}")
print("✅ All libraries loaded successfully — NO ERRORS")
```

```python
# Load CSV File
df = pd.read_csv("StudentFiles.csv")
print(f"✅ Dataset Loaded | Rows: {df.shape[0]}, Columns: {df.shape[1]}")
df.head()
```

```python
# Encode Text → Numbers
course_map = {'BSIT': 1, 'BSit': 1}
year_map = {'1st':1, '2nd':2, '3rd':3, '4th':4}
sem_map = {'1st':1, '2nd':2}

df['Course'] = df['Course'].map(course_map)
df['Year_Level'] = df['Year_Level'].map(year_map)
df['Semester'] = df['Semester'].map(sem_map)

# Fill missing values
df = df.fillna(df.mean(numeric_only=True))
print("✅ Data preprocessing complete")
df.head()
```

```python
# --------------------------
# CLASSIFICATION PROBLEM
# Predict Payment Status: 0 = Not Paid, 1 = Paid
# --------------------------

# Features & Target
X_class = df[['Course', 'Year_Level', 'Semester', 'Department_Count', 'Report_Mat', 'Prev.Avg_Approval']]
y_class = df['Payment_Status']

# Split Data: 80% Train, 20% Test
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(X_class, y_class, test_size=0.2, random_state=42)

# Train Model
class_model = LogisticRegression(max_iter=1000)
class_model.fit(X_train_c, y_train_c)

# Predict
y_pred_c = class_model.predict(X_test_c)

# Evaluation Metrics
print(f"✅ Classification Accuracy: {accuracy_score(y_test_c, y_pred_c):.4f}")
print("\n📊 Confusion Matrix:")
print(confusion_matrix(y_test_c, y_pred_c))
print("\n📋 Classification Report:")
print(classification_report(y_test_c, y_pred_c))

# Save Model
joblib.dump(class_model, "classification_model.pkl")
print("✅ Classification Model Saved")
```

```python
# --------------------------
# REGRESSION PROBLEM
# Predict Total Processing Days (Continuous Value)
# --------------------------

# Features & Target
X_reg = df[['Course', 'Year_Level', 'Semester', 'Payment_Status', 'Department_Count', 'Report_Mat', 'Prev.Avg_Approval']]
y_reg = df['Total_Days']

# Split Data
X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)

# Train Model
reg_model = LinearRegression()
reg_model.fit(X_train_r, y_train_r)

# Predict
y_pred_r = reg_model.predict(X_test_r)

# Evaluation Metrics
print(f"✅ MAE: {mean_absolute_error(y_test_r, y_pred_r):.4f}")
print(f"✅ MSE: {mean_squared_error(y_test_r, y_pred_r):.4f}")
print(f"✅ RMSE: {np.sqrt(mean_squared_error(y_test_r, y_pred_r)):.4f}")
print(f"✅ R² Score: {r2_score(y_test_r, y_pred_r):.4f}")

# Save Model
joblib.dump(reg_model, "regression_model.pkl")
print("✅ Regression Model Saved")
```
